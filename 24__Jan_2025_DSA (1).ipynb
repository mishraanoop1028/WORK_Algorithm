```python
import pandas as pd

# Load the Excel file
file_path = "SAADAA_assessment_data.xlsx"
df = pd.read_excel(file_path)

# Display the first few rows to understand the structure
print(df.head())
```

```python
# Calculate order frequency per customer
order_frequency = df.groupby('customer_id')['order_id'].count().reset_index()
order_frequency.columns = ['customer_id', 'order_frequency']

# Calculate total revenue per customer (Customer Lifetime Value)
customer_lifetime_value = df.groupby('customer_id')['order_value'].sum().reset_index()
customer_lifetime_value.columns = ['customer_id', 'lifetime_value']

# Merge the two metrics
customer_behavior = pd.merge(order_frequency, customer_lifetime_value, on='customer_id')

# Display the results
print("Customer Behavior Analysis:")
print(customer_behavior.head())
```

```python
# Top-performing products by revenue
top_products = df.groupby('product_sku')['order_value'].sum().sort_values(ascending=False).reset_index()
print("Top Performing Products:")
print(top_products.head())

# Seasonal trends (monthly revenue)
df['order_date'] = pd.to_datetime(df['order_date'])
df['month'] = df['order_date'].dt.month
monthly_revenue = df.groupby('month')['order_value'].sum().reset_index()
print("Monthly Revenue Trends:")
print(monthly_revenue)

# Revenue drivers (e.g., by product category)
revenue_by_category = df.groupby('product_category')['order_value'].sum().sort_values(ascending=False).reset_index()
print("Revenue by Product Category:")
print(revenue_by_category)
```

```python
# Calculate total conversions per marketing channel
conversions_by_source = df.groupby('source')['order_id'].count().reset_index()
conversions_by_source.columns = ['source', 'conversions']

# Calculate total revenue per marketing channel
revenue_by_source = df.groupby('source')['order_value'].sum().reset_index()
revenue_by_source.columns = ['source', 'total_revenue']

# Merge conversions and revenue
marketing_effectiveness = pd.merge(conversions_by_source, revenue_by_source, on='source')

# Display the results
print("Marketing Effectiveness:")
print(marketing_effectiveness)
```

```python
# Calculate return rates by product category and discount
return_rates = df.groupby(['product_category', 'discount_applied'])['return_status'].mean().reset_index()
return_rates.columns = ['product_category', 'discount_applied', 'return_rate']

# Display the results
print("Return Rates by Product Category and Discount:")
print(return_rates)
```

```python
# Calculate total quantity sold per product (assuming 'order_value' represents quantity)
product_sales = df.groupby('product_sku')['order_value'].sum().reset_index()
product_sales.columns = ['product_sku', 'total_quantity_sold']

# Identify products with low stock (assuming 'inventory_status' indicates stock levels)
if 'inventory_status' in df.columns:
    # Convert 'inventory_status' to a numeric value (e.g., 'In Stock' = 1, 'Out of Stock' = 0)
    df['inventory_status_numeric'] = df['inventory_status'].apply(lambda x: 1 if x == 'In Stock' else 0)

    # Calculate average inventory status per product
    product_stock = df.groupby('product_sku')['inventory_status_numeric'].mean().reset_index()
    product_stock.columns = ['product_sku', 'average_stock_status']

    # Merge sales and stock data
    inventory_risk = pd.merge(product_sales, product_stock, on='product_sku')
    inventory_risk['stockout_risk'] = inventory_risk['total_quantity_sold'] / (inventory_risk['average_stock_status'] + 1e-6)  # Avoid division by zero
    print("Products with High Stockout Risk:")
    print(inventory_risk.sort_values(by='stockout_risk', ascending=False).head())
else:
    print("No 'inventory_status' column found in the dataset.")

# Predictive model for inventory optimization (simplified example)
from sklearn.linear_model import LinearRegression

# Prepare data for prediction
inventory_data = df.groupby('product_sku').agg({'order_value': 'sum', 'inventory_status_numeric': 'mean'}).reset_index()
inventory_data.columns = ['product_sku', 'total_quantity_sold', 'average_stock_status']

# Fit a linear regression model
X = inventory_data[['total_quantity_sold']]
y = inventory_data['average_stock_status']
model = LinearRegression()
model.fit(X, y)

# Predict optimal stock levels
inventory_data['predicted_stock'] = model.predict(X)
print("Predicted Optimal Stock Levels:")
print(inventory_data[['product_sku', 'total_quantity_sold', 'average_stock_status', 'predicted_stock']])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Excel file
file_path = "SAADAA_assessment_data.xlsx"
df = pd.read_excel(file_path)

# Convert 'order_date' to datetime format (with error handling)
df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce')

# Check for missing values in 'order_date'
if df['order_date'].isna().any():
    print("Warning: Some dates could not be parsed and were set to NaT.")
    print(df[df['order_date'].isna()])

# Drop rows with invalid dates (optional)
df = df.dropna(subset=['order_date'])

# Extract hour from 'order_date'
df['hour'] = df['order_date'].dt.hour

# Categorize hours into time periods
def get_time_period(hour):
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 21:
        return 'Evening'
    else:
        return 'Night'

df['time_period'] = df['hour'].apply(get_time_period)

# Aggregate data by time period to calculate total sales (order_value) per time period
time_period_sales = df.groupby(['time_period'])['order_value'].sum().reset_index()

# Plot time-of-day trends
plt.figure(figsize=(10, 6))
sns.barplot(x='time_period', y='order_value', data=time_period_sales, palette='viridis', order=['Morning', 'Afternoon', 'Evening', 'Night'])
plt.title('Sales Trends by Time of Day')
plt.xlabel('Time Period')
plt.ylabel('Total Sales (Order Value)')
plt.show()

# Optional: Plot trends over time (e.g., daily trends by time period)
df['date'] = df['order_date'].dt.date
daily_time_period_sales = df.groupby(['date', 'time_period'])['order_value'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(x='date', y='order_value', hue='time_period', data=daily_time_period_sales, palette='viridis')
plt.title('Daily Sales Trends by Time of Day')
plt.xlabel('Date')
plt.ylabel('Total Sales (Order Value)')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend(title='Time Period')
plt.show()
```

```python
# Calculate total quantity sold per product
product_sales = df.groupby('Product')['Quantity'].sum().reset_index()
product_sales.columns = ['Product', 'Total Quantity Sold']

# Identify products with low stock (assuming there's a 'Stock' column)
if 'Stock' in df.columns:
    product_stock = df.groupby('Product')['Stock'].mean().reset_index()  # Average stock per product
    inventory_risk = pd.merge(product_sales, product_stock, on='Product')
    inventory_risk['Stockout Risk'] = inventory_risk['Total Quantity Sold'] / inventory_risk['Stock']
    print("Products with High Stockout Risk:")
    print(inventory_risk.sort_values(by='Stockout Risk', ascending=False).head())
else:
    print("No 'Stock' column found in the dataset.")

# Predictive model for inventory optimization (simplified example)
from sklearn.linear_model import LinearRegression

# Prepare data for prediction
inventory_data = df.groupby('Product').agg({'Quantity': 'sum', 'Stock': 'mean'}).reset_index()
inventory_data.columns = ['Product', 'Total Quantity Sold', 'Average Stock']

# Fit a linear regression model
X = inventory_data[['Total Quantity Sold']]
y = inventory_data['Average Stock']
model = LinearRegression()
model.fit(X, y)

# Predict optimal stock levels
inventory_data['Predicted Stock'] = model.predict(X)
print("Predicted Optimal Stock Levels:")
print(inventory_data[['Product', 'Total Quantity Sold', 'Average Stock', 'Predicted Stock']])
```
